# Bar Plot

```python
sns.barplot(x= data ['SL'],y= data ['SW'],palette='PiYG',errorbar=None)
plt.show()
```

# Heat Map

```python
corr_matrix = data.corr(numeric_only=True)
sns.heatmap(corr_matrix,annot=True,cmap='PiYG')
```


# Scatter Plot


```python
colors={'Iris-setosa':'red','Iris-versicolor':'blue','Iris-virginica':'yellow'}
plt.scatter(data['PL'], data['PW'],c=data['Classification'].map(colors))
plt.xlabel("PL")
plt.ylabel("PW")
plt.title("Petal Length vs Petal Width")

plt.show()
```

# Box Plot

```python
colors={'Iris-setosa':'red','Iris-versicolor':'blue','Iris-virginica':'yellow'}
plt.boxplot([data['SL'],data['SW'],data['PL'],data['PW']])
plt.show()
```
